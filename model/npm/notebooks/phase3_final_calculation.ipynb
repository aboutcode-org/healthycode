```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score

file_path = 'relevant_metrics_project.csv'
df = pd.read_csv(file_path)


x = df.drop(columns=['clean_url', 'Result'])
y = df['Result']

# split the dataset into training and testing sets
# 'test_size=0.25' means 25% of the data will be used to test the model
# 'stratify=y' ensures balanced distribution of Healthy/Unhealthy in both sets
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.25, random_state=42, stratify=y
)
```

```python
# 
model = LogisticRegression(max_iter=1000)
model.fit(x_train, y_train)

# Make predictions on the test set
y_pred = model.predict(x_test)

# Evaluate the model performance
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy:.2%}\n")
print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
# Trying with penalty='l2' handles multicollinearity by shrinking coefficients together.
model = LogisticRegression(l1_ratio=0, C=1.0, solver='lbfgs', max_iter=1000)
model.fit(x_train, y_train)

# Predict
y_pred = model.predict(x_test)

# Evaluate

print(f"Regularized Model Accuracy: {accuracy_score(y_test, y_pred):.2%}\n")
print("Classification Report:")
print(classification_report(y_test, y_pred))
```
